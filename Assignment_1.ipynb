# Direction Logic Implementation


```python
class Maze:
    def __init__(self, rows, cols):
        self.rows = rows
        self.cols = cols


def DFS(m):
    # Initialization
    start = (m.rows, m.cols)  # can be 4x4 or 5x5
    explored = [start]
    frontier = [start]
    
    print("original frontier=", frontier)
    print("original explored=", explored)
    
    dfsPath = {}
    
    # Continue loop as long as frontier is not empty
    while len(frontier) > 0:
        currCell = frontier.pop()
        
        # If current cell == goal, end the search
        if currCell == (1, 1):
            break
        
        # Explore all 4 directions in the order "ESNW" (East, South, North, West)
        for direction in ['E', 'S', 'N', 'W']:
            if direction == 'E':
                nextCell = (currCell[0], currCell[1] + 1)
            elif direction == 'S':
                nextCell = (currCell[0] + 1, currCell[1])
            elif direction == 'N':
                nextCell = (currCell[0] - 1, currCell[1])
            elif direction == 'W':
                nextCell = (currCell[0], currCell[1] - 1)
            
            # Check if nextCell is within the grid boundaries
            if 1 <= nextCell[0] <= m.rows and 1 <= nextCell[1] <= m.cols:
                if nextCell not in explored:
                    frontier.append(nextCell)
                    explored.append(nextCell)
                    dfsPath[nextCell] = currCell
    
    print("=============================================")
    # Print the contents of dfsPath
    print("{" + "\n".join("{!r}: {!r},".format(k, v) for k, v in dfsPath.items()) + "}")
    
    fwdPath = {}
    cell = (1, 1)
    
    print("=============================================")
    # Construct fwdPath to show the forward path from start to goal
    while cell != start:
        fwdPath[dfsPath[cell]] = cell
        cell = dfsPath[cell]
    
    print("=============================================")
    print("{" + "\n".join("{!r}: {!r},".format(k, v) for k, v in fwdPath.items()) + "}")
    
    return fwdPath
```

### Testing


```python
# Create a maze object with 5 rows and 5 columns
maze = Maze(5, 5)

# Call the DFS function with the maze object
fwdPath = DFS(maze)

# Print the forward path from start to goal
print(fwdPath)
```

#### dfsPath Dictionary:

This dictionary shows the path taken during the DFS.
Each key is a cell that was explored, and the value is the cell from which it was reached.

For example, (2, 1): (1, 1) means that from cell (2, 1), the next cell in the path is (1, 1).
